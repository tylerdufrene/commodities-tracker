```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

tickers = {
    "Natural Gas": "NG=F",
    "Feeder Cattle": "GF=F"
}

end = datetime.today()
start = end - timedelta(days=90)  # 3 months history

data = {}

for name, symbol in tickers.items():
    df = yf.download(symbol, start=start.strftime('%Y-%m-%d'), end=end.strftime('%Y-%m-%d'))
    df['% Change'] = df['Close'].pct_change() * 100
    data[name] = df

# Display last 5 rows of each
for name, df in data.items():
    print(f"\n{name} - Last 5 Days")
    print(df[['Close', 'Volume', '% Change']].tail())
```

```python
import matplotlib.pyplot as plt

for name, df in data.items():
    df['Close'].plot(title=f"{name} Price - Last 3 Months", figsize=(10, 4))
    plt.ylabel("Price")
    plt.grid()
    plt.show()
```
